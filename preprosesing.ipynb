```python
#CEK APAKAHN ADA DATA YG NULL
null_data = df.isnull().sum()
print("Kolom dengan nilai null:")
print(null_data[null_data > 0])
```

```python
#NAH TERNYATA ADA YG NULL
null_rows = df[df['overview'].isnull()]
print(null_rows)
```

```python
#KARENA ADA OVERVIEW YG TIDAK TERSEDIA daripada dibuang better kita iisi tidak ada 
df['overview'].fillna("Tidak tersedia", inplace=True)
```

```python
# Mengecek apakah masih ada nilai null di dataset
if df.isnull().sum().sum() == 0:
    print("Tidak ada nilai null di dataset.")
else:
    print("Masih ada nilai null di dataset.")
```
